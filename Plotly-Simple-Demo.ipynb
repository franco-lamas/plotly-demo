```python
import pandas as pd
dataAL30=pd.read_excel("AL30_history.xlsx")
```

```python
fig=px.line(dataAL30, x="Date",y="Bid")
fig.update_layout(template="plotly_dark",plot_bgcolor="#1C1C1C",margin=dict(l=2, r=80, t=2, b=20),paper_bgcolor="#151516")
fig.update_yaxes(gridcolor = '#333333',linecolor = '#636363',dtick = 2,side="right",title="Price USD")
fig.update_xaxes(gridcolor = '#333333',title="",showticklabels = True,dtick = 15*86400000,ticklabelmode="instant")
fig.update_traces(line=dict(color="#f68e26"))


fig.update_xaxes(rangeselector=dict(
                buttons=list([
                    dict(count=1,
                        label='1m',
                        step='month',
                        stepmode='backward'),
                    dict(count=6,
                        label='6m',
                        step='month',
                        stepmode='backward'),
                    dict(count=1,
                        label='YTD',
                        step='year',
                        stepmode='todate'),
                    dict(count=1,
                        label='1y',
                        step='year',
                        stepmode='backward'),
                    dict(count=2,
                        label='2y',
                        step='year',
                        stepmode='backward'),
                    dict(count=5,
                        label='5y',
                        step='year',
                        stepmode='backward'),
                    dict(count=10,
                        label='10y',
                        step='year',
                        stepmode='backward'),
                    dict(count=20,
                        label='20y',
                        step='year',
                        stepmode='backward'),
                        dict(step='all')
                ])
            ),
            rangeslider=dict(
                visible = True
            ),
            rangeselector_bgcolor="#2b2b2b")

fig.show()
```
